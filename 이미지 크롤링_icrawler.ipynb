* icrawler는 검색엔진에서 쉽게 이미지를 크롤링하는 패키지
* 참고로 Bing crawler만 잘 됨.

```python
pip install icrawler
```

```python
# 구글 크롤러가 안 되는 이유는 아래 사진 첨부함. 확실한지는 모름. 해결법 찾으면 공유좀..
from icrawler.builtin import BaiduImageCrawler, BingImageCrawler, GoogleImageCrawler
google_crawler = GoogleImageCrawler(
feeder_threads=1,
parser_threads=1,
downloader_threads=4,
storage={'root_dir': 'my_images_dir'})  #긁어올 이미지를 저장할 위치 지정
```

```python
Filters = dict(
size='large',
color= 'orange',
license = 'commercial,modify',
date=((2019,7,1),(2020,3,31)))
```

```python
google_crawler.crawl(keyword='shoes ads copy', offset=0, max_num=1000,
                     min_size=(200,200), filters=None,max_size=None, file_idx_offset=0)
```



* 출처: https://brunch.co.kr/@needleworm/52

```python
# bing 크롤러는 잘 됨
bing_crawler = BingImageCrawler(
downloader_threads=4,
storage={'root_dir': 't-shirts'})  # 저장할 폴더 이름 지정 't-shirts' 부분을 다른 폴더 이름으로 바꾸면 됨
```

```python
bing_crawler.crawl(keyword='t-shirts',filters=None, offset=0, max_num=1000)  # keyword부분에 어떤 검색어로 이미지를 검색할지 집어넣으면 됨. max_num으로 다운받을 사진 개수 지정 가능
```
